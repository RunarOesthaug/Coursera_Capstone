```python
import pandas as pd

df = pd.read_csv("postal_code.csv")
```

```python
df = df[["Postcode", "Borough", "Neighbourhood"]]
df
```

### Try to use geocoder to extract the coordinates for a given postal code

```python
import geocoder
print(geocoder.google('{}, Toronto, Ontario'.format(df)))
# doesn't work
```

### Instead use the provided coordinates

```python
coords = pd.read_csv("Geospatial_coordinates.csv")
coords.head()
```

```python
# Add Latitude and Longitude from coords to the Postal_code DataFrame
coords = coords.set_index("Postal Code")
coords = coords.reindex(index=df["Postcode"])
coords = coords.reset_index()
df[["Latitude", "Longitude"]] = coords[["Latitude", "Longitude"]]
df
```

### Use folium to display a map of Toronto

```python
import folium
Latitude = 43.68
Longitude = -79.35
map_Toronto = folium.Map(location=[Latitude, Longitude], zoom_start=11)
map_Toronto
```

### Add markers to each Neighborhood

```python
# add markers to map
for lat, lng, label in zip(df['Latitude'], df['Longitude'], df['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=0.3,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.5,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
